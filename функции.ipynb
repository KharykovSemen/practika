```python
def power(base, exp):
    return base ** exp

print(power(2, 3))
print(power(5, 0)) 
print(power(10, 2)) 
```

```python
def apply_to_list(func, items):
    return [func(item) for item in items]

def square(x):
    return x ** 2

numbers = [1, 2, 3, 4, 5]

squared_numbers = apply_to_list(square, numbers)
print(squared_numbers) 
```

```python
def make_adder(n):
    def adder(x):
        return x + n
    return adder

add_five = make_adder(5)
print(add_five(10))  

add_ten = make_adder(10)
print(add_ten(3))    
```

```python
import time

def timeit(func):
    def wrapper(*args, **kwargs):
        start_time = time.time()  # время начала выполнения
        result = func(*args, **kwargs)  # вызов функции
        end_time = time.time()  # время окончания выполнения
        print(f"Функция '{func.__name__}' выполнена за {end_time - start_time:.6f} секунд")
        return result
    return wrapper

@timeit
def multiplication(a, b):
    return a * b

result = multiplication(123456789, 9999)
print("Результат умножения:", result)
```

```python
import time

def timeit(func):
    def wrapper(*args, **kwargs):
        start_time = time.time()
        result = func(*args, **kwargs)
        end_time = time.time()
        print(f"Время выполнения '{func.__name__}': {end_time - start_time:.6f} секунд")
        return result
    return wrapper

@timeit
def process_numbers(*args, **kwargs):
    total_args = sum(args)
    multiplier = kwargs.get('multiplier', 1)
    result = total_args * multiplier
    return result

result1 = process_numbers(1, 2, 3, 4, 5, multiplier=10)
print("Результат:", result1)

result2 = process_numbers(10, 20, 30)
print("Результат:", result2)
```
