```python
# Ваш код здесь
def even(a, b):
  for num in range(a, b):
        if num % 2 == 0:
            yield num

# Пример проверки:
print(list(even(3, 12)))  # Ожидаем: [4, 6, 8, 10]
```

```python
# Ваш код здесь
nums = list(range(1, 11))
# Результат положите в переменную result
result = [(lambda x: x ** 2)(x) for x in nums if x % 2 != 0]
print(result)  # Ожидаем: [1, 9, 25, 49, 81]
```

```python
# Ваш код здесь
def flatten(iterable):
    for sublist in iterable:
        yield from sublist


print(list(flatten([[1,2],[3],[4,5,6]])))  # Ожидаем: [1,2,3,4,5,6]
```

```python
# Подсказка: используйте генераторы и стандартные инструменты
# (например, str.isalnum, itertools, collections.Counter)

from collections import Counter
from itertools import chain

text = "Python, python! Генераторы и lambda: быстрее, компактнее и удобнее. Python? Да, Python."
N = 3

# Ваш код здесь
def normalize(s):
      return ''.join(c.lower() if c.isalnum() or c.isspace() else ' ' for c in s)


def words(s):
    return normalize(s).split()

# Основной пайплайн
def pipeline(text, n):
    # Цепочка генераторов
    word_sequence = words(text)
    frequencies = Counter(word_sequence)
    top_n = frequencies.most_common(n)
    return top_n

# Запуск пайплайна
top = pipeline(text, N)
print(top)

# Подсчёт и вывод
# top = ...
# print(top)  # Ожидаем топ-3 слов (например, [('python', 3), ('и', 1), ...])
```
